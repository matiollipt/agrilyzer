# Visualization Tests

```python
import pandas as pd
from agrilyzer.analyzers import Agrilyzer
from agrilyzer.visualizers import Visualizer
```

## Instantiate AidWeather client and Visualizer


```python
aw = Agrilyzer("TestSite", -12.5, -49.2, "2023-01-01", "2023-12-31")
viz = Visualizer(aw)
```

## Load agro data

```python
df_agro = pd.read_csv("sample.csv")
df_agro
```

## Plot weather data only (moving average of 7 days)

```python
viz.plot(
    cols=["T2M", "PRECTOTCORR"], ma=7, dual=True, apply_theme="ggplot"
)
```

## Plot agro and weather overlay

```python
viz.overlay(
    other=df_agro,
    other_cols=["height_cm"],
    weather_cols=["T2M"],
    apply_theme="ggplot",
)
```
